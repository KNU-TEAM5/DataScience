```python
# 인코딩 자동 시도
def read_csv_safely(path):
    for enc in ["utf-8-sig", "utf-8", "cp949", "euc-kr"]:
        try:
            return pd.read_csv(path, encoding=enc)
        except Exception:
            pass
    # 마지막 시도: 기본
    return pd.read_csv(path)

# '오전/오후' 또는 AM/PM 형식 모두 처리
def parse_time_kor(value):
    s = str(value).strip()
    # '오전/오후'를 AM/PM으로 치환
    s = s.replace("오전", "AM").replace("오후", "PM")
    # 다양한 포맷 시도
    fmts = ["%p %I:%M:%S", "%I:%M:%S %p", "%H:%M:%S"]
    for fmt in fmts:
        try:
            return datetime.datetime.strptime(s, fmt).time()
        except Exception:
            continue
    # 실패시 None
    return None

# 컬럼명 정규화(공백 제거, 소문자)
def normalize_columns(df):
    df = df.copy()
    df.columns = (
        pd.Series(df.columns)
        .astype(str)
        .str.strip()
        .str.replace(r"\s+", "_", regex=True)
        .str.replace(r"[^\w]+", "_", regex=True)
        .str.lower()
    )
    return df
```

```python
def parse_time_kor(series):
    # "오전/오후" → "AM/PM"으로 바꾸고 datetime 파싱
    s = series.astype(str).str.replace("오전", "AM").str.replace("오후", "PM")
    return pd.to_datetime(s, format="%p %I:%M:%S.%f", errors="coerce").dt.time

frames = []
for f in sensor_csvs:
    df = read_csv_safely(f)
    df = normalize_columns(df)

    # 컬럼 이름 표준화
    lot_col = "lot" if "lot" in df.columns else None
    date_str = re.search(r"(\d{4})\.(\d{2})\.(\d{2})", f)
    if date_str:
        y, mth, d = map(int, date_str.groups())
        df["date"] = f"{y:04d}-{mth:02d}-{d:02d}"
    else:
        df["date"] = pd.Timestamp.today().strftime("%Y-%m-%d")

    # Time 파싱
    df["time_parsed"] = parse_time_kor(df["time"])
    df["timestamp"] = pd.to_datetime(df["date"] + " " + df["time_parsed"].astype(str), errors="coerce")

    keep_cols = ["timestamp","date","time","ph","temp","current"]
    if lot_col: keep_cols.append(lot_col)
    frames.append(df[keep_cols])

dd = pd.concat(frames, ignore_index=True)
dd = dd.sort_values("timestamp").dropna(subset=["timestamp"]).reset_index(drop=True)
dd = dd.set_index("timestamp")
dd.head()
```

1. 히스토그램들 (pH, Temp, Current)

각각 센서 값의 분포를 보여주는 그래프

pH
- 대부분 9.5~11 근처에 몰려 있음. → 공정에서 약품(pH 조절제)이 일정하게 관리되고 있다는 의미.
- 몇 개 값이 12 근처로 튀어 있음 → 과투입이나 센서 이상 가능.

Temp (온도)
- 44~48℃ 구간에 주로 분포. → 공정이 설정된 온도 범위에서 안정적으로 운전되는 모습.
- 꼬리 부분(52℃ 이상)이 있으면, 열 교환기 문제나 반응 과열 징후일 수 있음.

Current (전류)
- 7~9A 근처에 집중. → 장비(전해조 등)가 일정 부하에서 작동.
- 6A 이하로 떨어진 구간은 장비 부하 저하, 공급 문제, 순간적인 셧다운 가능성을 의미.

👉 인사이트:
히스토그램은 정상 범위와 이탈 구간을 직관적으로 보여줍니다. 제조 현장에서는 이걸 기준으로 “관리 허용 범위(spec range)”를 설정하고, 범위를 벗어난 값을 불량 후보로 관리합니다.

히스토그램 → 각 변수의 정상 분포 범위를 확인, 튀어나온 값이 있으면 공정 이상 후보.

2. 상관관계 히트맵

pH–Temp, Temp–Current 등 모든 상관계수가 0.0~0.02 수준으로 매우 낮음.

즉, 세 변수(pH, Temp, Current)는 서로 거의 독립적으로 움직인다는 뜻.

👉 인사이트:

만약 Temp가 올라가면 Current도 올라간다(양의 상관) 같은 패턴이 있어야 정상인데, 여기서는 거의 없음 → 공정 변수들이 독립적으로 제어되고 있다고 해석 가능.

이는 장비 설계상 의도일 수도 있고, 반대로 센서 간 상관이 약해져서 공정 제어 로직이 제대로 연동되지 않는 신호일 수도 있음.

상관관계 히트맵 → 변수들이 독립적으로 움직이는지, 아니면 연동성이 있는지를 보여줌. (연동이 깨지면 공정 이상)

```python
# === 8) 강화 라벨링: Error Lot 매칭 진단 + 자동 보정 ===
import os, re, numpy as np, pandas as pd

def read_csv_safely(path):
    for enc in ["utf-8-sig","utf-8","cp949","euc-kr"]:
        try:
            return pd.read_csv(path, encoding=enc)
        except Exception:
            pass
    return pd.read_csv(path)

def normalize_columns(df):
    df = df.copy()
    df.columns = (pd.Series(df.columns).astype(str)
                  .str.strip().str.replace(r"\s+","_",regex=True)
                  .str.replace(r"[^\w]+","_",regex=True)
                  .str.lower())
    return df

def to_date_str(x):
    if pd.isna(x): return np.nan
    s = str(x).strip()
    s = (s.replace("년","-").replace("월","-").replace("일","")
           .replace("/","-").replace(".","-"))
    dt = pd.to_datetime(s, errors="coerce")
    return np.nan if pd.isna(dt) else dt.strftime("%Y-%m-%d")

def norm_lot(x):
    if pd.isna(x): return np.nan
    s = str(x).strip()
    s = re.sub(r"[\s\-_\/]+","", s)
    return s.upper()

# 1) 에러 CSV 찾기
error_csvs = [f for f in os.listdir() if f.lower().endswith(".csv") and "error" in f.lower()]
if not error_csvs:
    raise FileNotFoundError("Error Lot list.csv 파일을 찾지 못했습니다.")
error_csv = error_csvs[0]

error = read_csv_safely(error_csv)
raw_error_cols = list(error.columns)
error = normalize_columns(error)

# 2) 날짜/LOT 후보 탐색
date_candidates = [c for c in error.columns if any(k in c for k in ["date","날짜","일자","dt"])]
lot_candidates  = [c for c in error.columns if "lot" in c]

if not date_candidates:
    raise KeyError(f"에러 파일에 날짜 컬럼이 없습니다. 원본 컬럼: {raw_error_cols}")

date_col = date_candidates[0]
error["date_norm"] = error[date_col].map(to_date_str)

lots_norm_cols = []
for lc in lot_candidates:
    nc = lc + "_norm"
    error[nc] = error[lc].map(norm_lot)
    lots_norm_cols.append(nc)

# 3) 센서 쪽도 통일
if "date" not in dd.columns:
    dd["date"] = dd.index.strftime("%Y-%m-%d")
else:
    dd["date"] = dd["date"].map(to_date_str)

sensor_lot_cols = [c for c in dd.columns if "lot" in c.lower()]
if sensor_lot_cols:
    sensor_lot_col = sensor_lot_cols[0]
    dd["lot_norm"] = dd[sensor_lot_col].map(norm_lot)
else:
    dd["lot_norm"] = np.nan

# 4) 진단 출력
sensor_dates = set(dd["date"].dropna())
error_dates  = set(error["date_norm"].dropna())
overlap_dates = sorted(sensor_dates & error_dates)

print("[진단] sensor 날짜 수:", len(sensor_dates), "| error 날짜 수:", len(error_dates))
print("      교집합 날짜 수:", len(overlap_dates))
print("      교집합 날짜 예시:", overlap_dates[:10])

if sensor_lot_cols and lots_norm_cols:
    print("sensor LOT 예시:", dd["lot_norm"].dropna().astype(str).unique()[:5])
    ex_err_lots = pd.concat([error[c] for c in lots_norm_cols], axis=1).stack().dropna().astype(str).unique()[:5]
    print("error LOT 예시:", ex_err_lots)

# 5) 1차 라벨(date+lot), 0개면 2차(date-only)로 폴백
def label_with(mode="date_lot"):
    if mode == "date_lot" and sensor_lot_cols and lots_norm_cols:
        err_pairs = set()
        for c in lots_norm_cols:
            sub = error[["date_norm", c]].dropna()
            err_pairs |= set(zip(sub["date_norm"], sub[c]))
        qc = dd.apply(lambda r: 0 if (r["date"], r["lot_norm"]) in err_pairs else 1, axis=1).astype(int)
        used = "date_lot"
    else:
        err_dates = set(error["date_norm"].dropna())
        qc = dd["date"].apply(lambda d: 0 if d in err_dates else 1).astype(int)
        used = "date_only"
    return qc, used

dd["QC"], mode_used = label_with("date_lot")
vc = dd["QC"].value_counts().to_dict()
print(f"\nQC 분포(1차 {mode_used}):", vc)

if vc.get(0,0) == 0:  # 불량이 0개면 날짜만으로 재시도
    dd["QC"], mode_used = label_with("date_only")
    vc = dd["QC"].value_counts().to_dict()
    print(f"QC 분포(2차 {mode_used}):", vc)

# 최종 결과 출력
print("\n최종 QC 분포:", dd["QC"].value_counts().to_dict())
```

```python
# === 10) 지도 ↔ 비지도 자동 전환 ===
import numpy as np
import pandas as pd

# 1) 피처 선택: 표준화 컬럼 우선 사용, 없으면 원래 이름으로 폴백
cand_order = [["ph_std","temp_std","current_std"], ["ph","temp","current"]]
features = None
for trio in cand_order:
    cols = [c for c in trio if c in dd.columns]
    if cols:
        features = cols; break
if not features:
    raise ValueError(f"센서 피처를 찾지 못했습니다. dd.columns={list(dd.columns)[:30]}")

# 2) X 구성 (숫자화 + 결측 제거)
X = dd[features].apply(pd.to_numeric, errors="coerce")
mask = X.notna().all(axis=1)
X = X.loc[mask]

# 3) 지도 여부 결정: QC가 있고 클래스가 2개 이상이면 지도
use_supervised = ("QC" in dd.columns) and (dd.loc[X.index, "QC"].dropna().nunique() >= 2)

if use_supervised:
    # ===== 지도학습(분류) 경로 =====
    print("== 지도학습(분류) 경로: XGBoost ==")
    from sklearn.model_selection import train_test_split
    from xgboost import XGBClassifier

    y = dd.loc[X.index, "QC"].astype(int)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, shuffle=True, random_state=42, stratify=y
    )

    clf = XGBClassifier(
        n_estimators=300, max_depth=5, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        reg_lambda=1.0, random_state=42, eval_metric="logloss"
    )
    clf.fit(X_train, y_train)

    y_pred  = clf.predict(X_test)
    y_score = clf.predict_proba(X_test)[:, 1]  # 양성(불량=1) 확률

else:
    # ===== 비지도(이상치 탐지) 경로 =====
    print("== 비지도(Anomaly) 경로: IsolationForest ==")
    from sklearn.ensemble import IsolationForest

    iso = IsolationForest(
        n_estimators=300, contamination=0.05,  # 필요시 0.01~0.10로 조정
        random_state=42
    )
    iso.fit(X)

    # score_samples는 값이 클수록 정상 → 뒤집어서 "클수록 이상"으로 사용
    scores = -iso.score_samples(X)
    thr = np.quantile(scores, 0.95)  # 상위 5% 이상치
    flags = (scores >= thr).astype(int)

    # X와 동일 인덱스의 결과 프레임
    Xc = X.copy()
    Xc["anomaly_score"] = scores
    Xc["anomaly_flag"]  = flags

    # dd와 병합: 인덱스(timestamp) 중복을 고려하여 reset_index 후 concat
    out = dd.loc[X.index].reset_index().copy()  # timestamp가 컬럼으로 풀림
    out = pd.concat([out, Xc[["anomaly_score","anomaly_flag"]].reset_index(drop=True)], axis=1)

    # 상위 이상치 10개 미리보기
    print("상위 10개 이상치 샘플:")
    display(out.sort_values("anomaly_score", ascending=False).head(10))

    # 분포 확인
    import matplotlib.pyplot as plt
    plt.figure(figsize=(5,3))
    plt.hist(scores, bins=50)
    plt.title("Anomaly Score Distribution")
    plt.xlabel("score (higher = more anomalous)")
    plt.ylabel("count")
    plt.show()

```

👉 인사이트:

오른쪽 꼬리(high score 영역)는 공정 이상, 설비 이상, 혹은 불량 LOT 발생 구간일 가능성이 큼.

히스토그램만 보면 정상 데이터가 훨씬 많지만, tail에 소수의 고점이 있다는 게 중요한 신호.
